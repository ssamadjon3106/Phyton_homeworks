# scatter

```python
a=np.arange(50)
b=np.random.randint(0, 50, 50)
sizes=np.abs(np.random.randn(50))*100
plt.scatter(a,b, c=b, s=sizes, cmap='cool')
plt.show()
```

```python
x=['red', 'blue', 'green', 'orange']
y=[8, 9, 2, 54]

plt.bar(x, y, color=[c.lower() for c in x], width=0.6)
plt.title('muhabbat')
plt.xlabel('we')
plt.ylabel('measerement')
```

```python
cat=['A', 'd', 'f', 'w', 'g']
v1=[12, 23, 34, 45, 65]
v2=[32, 43, 45, 65, 67]

x=np.arange(len(cat))

plt.bar(x+0.1, v1, color='red', width=0.2 , label='hey')
plt.bar(x-0.1, v2,color='b', width=0.2 , label='uffff')
plt.xticks(x, cat)
plt.legend()
plt.show()
```

```python
cat=['A', 'd', 'f', 'w', 'g']
v1=[124, 23, 34, 45, 65]
v2=[32, 43, 45, 65, 67]

x=np.arange(len(cat))

plt.bar(cat, v1,  label='hey', color='r' , edgecolor='k')
plt.bar(cat, v2,label='uffff', color='b' ,edgecolor='k')

plt.legend()
plt.show()
```
