```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 4*x + 4

x = np.linspace(-10, 10, 400)
y = f(x)

plt.plot(x, y, label=r'$f(x) = x^2 - 4x + 4$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the function $f(x) = x^2 - 4x + 4$')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values from 0 to 2
x = np.linspace(0, 2 * np.pi, 400)

# Define sine and cosine values
sin_x = np.sin(x)
cos_x = np.cos(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, sin_x, label='sin(x)', color='blue', linestyle='-', marker='o', markersize=4)
plt.plot(x, cos_x, label='cos(x)', color='red', linestyle='--', marker='x', markersize=4)

# Customize the plot
plt.title("Sine and Cosine Functions")
plt.xlabel("x (radians)")
plt.ylabel("Function value")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values for each plot
x1 = np.linspace(-2, 2, 400)        # for cubic
x2 = np.linspace(-2*np.pi, 2*np.pi, 400)  # for sin
x3 = np.linspace(-2, 2, 400)        # for exp
x4 = np.linspace(0, 5, 400)         # for log(x+1), x >= 0

# Compute y values
y1 = x1**3
y2 = np.sin(x2)
y3 = np.exp(x3)
y4 = np.log(x4 + 1)

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Top-left: cubic
axs[0, 0].plot(x1, y1, color='blue')
axs[0, 0].set_title(r'$f(x) = x^3$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')
axs[0, 0].grid(True)

# Top-right: sine
axs[0, 1].plot(x2, y2, color='green')
axs[0, 1].set_title(r'$f(x) = \sin(x)$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')
axs[0, 1].grid(True)

# Bottom-left: exponential
axs[1, 0].plot(x3, y3, color='red')
axs[1, 0].set_title(r'$f(x) = e^x$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')
axs[1, 0].grid(True)

# Bottom-right: log(x+1)
axs[1, 1].plot(x4, y4, color='purple')
axs[1, 1].set_title(r'$f(x) = \log(x + 1)$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random points for x and y from uniform [0, 10]
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Different markers and colors to cycle through
markers = ['o', 's', '^', 'D', 'P', '*', 'X']
colors = plt.cm.rainbow(np.linspace(0, 1, len(markers)))

plt.figure(figsize=(8, 6))

# Plot each point with a different color and marker cycling through the lists
for i in range(100):
    plt.scatter(x[i], y[i], color=colors[i % len(colors)], marker=markers[i % len(markers)], edgecolor='black', s=80)

plt.title("Scatter Plot of 100 Random Points")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random values from a normal distribution (mean=0, std=1)
data = np.random.normal(0, 1, 1000)

# Plot histogram with 30 bins and adjusted transparency
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')

plt.title("Histogram of Normally Distributed Data (mean=0, std=1)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # necessary for 3d plotting

# Create grid of x, y values
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

# Compute f(x, y)
Z = np.cos(X**2 + Y**2)

# Create plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot surface with a colormap
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add colorbar
fig.colorbar(surf, shrink=0.5, aspect=10)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y) = cos(x + y)')
ax.set_title('3D Surface Plot of f(x, y) = cos(x + y)')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]

# Custom colors for each bar
colors = ['#FF5733', '#33C1FF', '#33FF57', '#FF33A8', '#FFC733']

plt.figure(figsize=(8, 6))
plt.bar(products, sales, color=colors)

plt.title('Sales Data for Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Time periods
time_periods = ['T1', 'T2', 'T3', 'T4']

# Sample data for categories
category_a = [5, 7, 3, 4]
category_b = [2, 4, 7, 6]
category_c = [3, 5, 2, 3]

# Positions on x-axis
x = np.arange(len(time_periods))

# Plot stacked bars
plt.figure(figsize=(8, 6))
plt.bar(x, category_a, label='Category A', color='skyblue')
plt.bar(x, category_b, bottom=category_a, label='Category B', color='salmon')
# For category C, bottom is sum of A + B for stacking
bottom_ab = np.array(category_a) + np.array(category_b)
plt.bar(x, category_c, bottom=bottom_ab, label='Category C', color='lightgreen')

# Labels and title
plt.xticks(x, time_periods)
plt.xlabel('Time Periods')
plt.ylabel('Values')
plt.title('Stacked Bar Chart of Categories over Time')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
