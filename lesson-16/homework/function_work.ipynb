```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 4*x + 4

x = np.linspace(-10, 10, 400)
y = f(x)

plt.plot(x, y, label=r'$f(x) = x^2 - 4x + 4$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the function $f(x) = x^2 - 4x + 4$')
plt.legend()
plt.grid(True)
plt.show()

```
