```python
import pandas as pd

# Load data files (adjust filenames if needed)
df_iris = pd.read_json('iris.json')
df_titanic = pd.read_excel('titanic.xlsx')
df_movie = pd.read_csv('movie.csv')
```

```python
df_iris.columns = [col.lower() for col in df_iris.columns]
iris_selected = df_iris[['sepallength', 'sepalwidth']]
iris_selected.head()


```

```python
# Assuming you already loaded the titanic.xlsx file into a DataFrame named df_titanic

# Filter passengers with age above 30
age_above_30 = df_titanic[df_titanic['Age'] > 30]

# Count number of male and female passengers in that filtered data
gender_counts = age_above_30['Sex'].value_counts()

print(gender_counts)
```

```python
# Assuming your movie.csv is loaded into df_movie

# Filter rows where duration > 120
long_movies = df_movie[df_movie['duration'] > 120]

# Sort by director_facebook_likes descending
sorted_movies = long_movies.sort_values(by='director_facebook_likes', ascending=False)

sorted_movies.head(10)  # Show top 10 rows
```

```python
iris_stats = df_iris.describe().loc[['mean', '50%', 'std']]
iris_stats.rename(index={'50%': 'median'}, inplace=True)
print("Iris statistics:")
print(iris_stats)
```

```python
age_min = df_titanic['Age'].min()
age_max = df_titanic['Age'].max()
age_sum = df_titanic['Age'].sum()
print("\nTitanic passenger age stats:")
print(f"Min age: {age_min}")
print(f"Max age: {age_max}")
print(f"Sum of ages: {age_sum}")
```

```python
director_likes = df_movie.groupby('director_name')['director_facebook_likes'].sum()
top_director = director_likes.idxmax()
top_likes = director_likes.max()

print(f"\nDirector with highest total Facebook likes: {top_director} ({top_likes} likes)")

# Find 5 longest movies and their directors
longest_movies = df_movie.nlargest(5, 'duration')[['movie_title', 'director_name', 'duration']]
print("\nTop 5 longest movies and their directors:")
print(longest_movies)
```
